Практическая работа 3
Тема: нечеткая логика, нечеткий вывод, алгоритм Мамдани
Дано: условие задачи (по вариантам).
Требуется:
Реализовать механизм логического вывода с помощью нечетких правил и нечетких
множеств.
Для решения предлагается использовать алгоритм Мамдани, обычно включающий в себя
следующие этапы:
1. Фаззификация - преобразование конкретных значений в вербальное нечеткое
представление.
2. Составление нечетких правил, представляющих рассуждение в нечетких (размытых)
терминах близких к естественному языку.
3. Аггрегация результатов вывода.
4. Дефаззификация — преобразование нечеткого вывода в конекретное значение.
Выполнить решение со следующими типами функций принадлежности:
1. Треугольная: fuz.trimf( distance.universe, [0, 0, 50] )
2. Трапецевидная: fuz.trapmf( distance.universe, [0, 0, 30, 50] )
3. Функция Гаусса: fuz.gaussmf( speed.universe, mean = 0, sigma = 15 )
Для реализации предлагается использовать язык Python и библиотеку для работы с
нечеткой логикой scikit-fuzzy.
Документация: https://pythonhosted.org/scikit-fuzzy/

```python
#Пример выполнения задачи (для треугольной функции принадлежности):
# 0. если требуется установить пакет
!pip install scikit-fuzzy
import numpy as np
import skfuzzy as fuz
from skfuzzy import control as ctrl
# 1. Определяем входные и выходные переменные
# Antecedent определяет логический посыл, переменные, на основе которых
# строится рассуждение
# Consequent определяет нечеткую выходную переменную — рузельтат
# np.arange(0,101,1) задает множество чисел 0,1,2,...,100 — таким способом
# задаем шкалу для нечеткой переменной
distance = ctrl.Antecedent( np.arange(0, 101, 1), "Расстояние" )
speed = ctrl.Antecedent( np.arange(0, 111, 1), "Скорость" )
brake = ctrl.Consequent( np.arange(0, 101, 1), "Торможение" )
# 2. Определение функций принадлежности входных переменных расстояния и
# скорости
distance['Близко'] = fuz.trimf(distance.universe, [0, 0, 50])
distance['Средне'] = fuz.trimf(distance.universe, [0, 50, 100])
distance['Далеко'] = fuz.trimf(distance.universe, [50, 100, 100])
distance.view()

speed["Медленная"] = fuz.trimf( speed.universe, [0, 0, 50] )
speed["Умеренная"] = fuz.trimf( speed.universe, [0, 50, 100] )
speed["Быстрая"] = fuz.trimf( speed.universe, [50, 110, 110] )
speed.view()

# 3. Функции принадлежности для выходной переменной «усилие торможения»
brake["Легкое"] = fuz.trimf( brake.universe, [0, 0, 50] )
brake["Умеренное"] = fuz.trimf( brake.universe, [0, 50, 100] )
brake["Жесткое"] = fuz.trimf( brake.universe, [50, 100, 100] )
brake.view()

# 4. правила вывода решения об усилии торможения на основании характеристик
# текущей скорости и расстояния до цели
rule1 = ctrl.Rule( distance["Близко"] & speed["Быстрая"], brake["Жесткое"] )
rule2 = ctrl.Rule( distance["Близко"] & speed["Умеренная"], brake["Умеренное"])
rule3 = ctrl.Rule(distance["Средне"] & speed["Быстрая"], brake["Умеренное"])
rule4 = ctrl.Rule(distance["Далеко"] & speed["Медленная"], brake["Легкое"])
rule5 = ctrl.Rule(distance["Средне"] & speed["Медленная"], brake["Легкое"])
# 5. система нечеткого управления на основе описанных правил
braking_ctrl = ctrl.ControlSystem( [rule1, rule2, rule3, rule4, rule5] )
braking = ctrl.ControlSystemSimulation( braking_ctrl )
# 6. Подаем на вход конкретные значения скорости и расстояния
braking.input["Расстояние"] = 30
braking.input["Скорость"] = 70
# 7. Вычисляем результат
braking.compute()
# 8. Выводим результат
print(f"Интенсивность торможения: { braking.output['Торможение']:.2f} %")
brake.view( sim = braking)
```

Задача: «Управление светофором на перекрестке»
Суть: Система должна регулировать длительность зеленого сигнала светофора в зависимости от
загруженности дороги, времени суток и наличия пешеходов.
Входные переменные:
• traffic_density — загруженность дороги:
◦ Низкая / Средняя / Высокая.
• time_of_day — время суток:
◦ Утро / День / Вечер / Ночь.
• pedestrians — количество пешеходов на переходе:
◦ Нет / Мало / Много.
Выходная переменная:
• green_light_duration — длительность зеленого сигнала (сек):
◦ Короткая / Средняя / Длинная.
Примеры правил:
• Если загруженность высокая и пешеходов нет, то зеленый длинный.
• Если загруженность средняя и пешеходов много, то зеленый короткий.
• Если время суток ночь и загруженность низкая, то зеленый короткий.
• Если время суток утро и загруженность высокая, то зеленый длинный.
• Если пешеходов мало и загруженность средняя, то зеленый средний.


```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

# Создание универсальных множеств
traffic_density = ctrl.Antecedent(np.arange(0, 11, 1), 'traffic_density')
time_of_day = ctrl.Antecedent(np.arange(0, 25, 1), 'time_of_day')
pedestrians = ctrl.Antecedent(np.arange(0, 11, 1), 'pedestrians')
green_light_duration = ctrl.Consequent(np.arange(0, 61, 1), 'green_light_duration')

# График функций принадлежности для каждой переменной
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Загруженность дорог
traffic_density['low'] = fuzz.trimf(traffic_density.universe, [0, 0, 5])
traffic_density['medium'] = fuzz.trimf(traffic_density.universe, [3, 5, 8])
traffic_density['high'] = fuzz.trimf(traffic_density.universe, [6, 10, 10])
traffic_density.view(ax=axes[0][0])
axes[0][0].set_title('Загруженность дороги')

# Время суток
time_of_day['morning'] = fuzz.trimf(time_of_day.universe, [0, 6, 10])
time_of_day['day'] = fuzz.trimf(time_of_day.universe, [8, 12, 16])
time_of_day['evening'] = fuzz.trimf(time_of_day.universe, [14, 18, 22])
time_of_day['night'] = fuzz.trimf(time_of_day.universe, [20, 24, 24])
time_of_day.view(ax=axes[0][1])
axes[0][1].set_title('Время суток')

# Количество пешеходов
pedestrians['none'] = fuzz.trimf(pedestrians.universe, [0, 0, 3])
pedestrians['few'] = fuzz.trimf(pedestrians.universe, [2, 5, 8])
pedestrians['many'] = fuzz.trimf(pedestrians.universe, [6, 10, 10])
pedestrians.view(ax=axes[1][0])
axes[1][0].set_title('Количество пешеходов')

# Длительность зеленого сигнала
green_light_duration['short'] = fuzz.trimf(green_light_duration.universe, [0, 0, 20])
green_light_duration['medium'] = fuzz.trimf(green_light_duration.universe, [15, 30, 45])
green_light_duration['long'] = fuzz.trimf(green_light_duration.universe, [40, 60, 60])
green_light_duration.view(ax=axes[1][1])
axes[1][1].set_title('Продолжительность зеленого сигнала')

plt.tight_layout()
plt.show()

# Правила вывода
rule1 = ctrl.Rule(traffic_density['high'] & pedestrians['none'], green_light_duration['long'])
rule2 = ctrl.Rule(traffic_density['medium'] & pedestrians['many'], green_light_duration['short'])
rule3 = ctrl.Rule(time_of_day['night'] & traffic_density['low'], green_light_duration['short'])
rule4 = ctrl.Rule(time_of_day['morning'] & traffic_density['high'], green_light_duration['long'])
rule5 = ctrl.Rule(pedestrians['few'] & traffic_density['medium'], green_light_duration['medium'])

# Контроллер
system_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5])
simulation = ctrl.ControlSystemSimulation(system_ctrl)

# Тестируем
input_data = {
    'traffic_density': 7,
    'time_of_day': 15,
    'pedestrians': 2
}

for key in input_data.keys():
    simulation.input[key] = input_data[key]

# Вычисляем результат
simulation.compute()
result = round(simulation.output["green_light_duration"], 2)

# Показываем результирующую кривую (график нечеткого вывода)
green_light_duration.view(sim=simulation)
plt.title(f'Результат расчета ({result} секунд)')
plt.show()

print(f'Длительность зеленого сигнала: {result} секунд.')
```
