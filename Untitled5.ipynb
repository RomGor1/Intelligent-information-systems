```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt

# Исходные данные для варианта №29
data = {
    "X": [
        1.35367822, 0.56204431, 1.33033853, 2.79569205, 2.74407345, 0.67715423,
        -0.26284334, 1.91179911, 2.10742948, 0.34586732, 3.29311281, 1.54572677,
        0.28490835, 2.50322374, 2.46560974, 2.75153077, 4.45520412, 1.65824123,
        2.66515022, 3.5870544, 2.38734622, 0.13798494, 1.26842946, 2.56952111,
        1.64844924, 3.77549143, 1.3221597, 2.19121461, 0.09522418, 0.39582865,
        1.42846394, 2.46569716, 1.09679571, 0.4086105, 2.32815239, 2.57821122,
        7.57533657, 7.20301008, 6.64312685, 7.18168084, 7.41996058, 7.13885368,
        6.72460861, 7.40989281, 7.8006125, 7.87775142, 7.62324613, 6.99926527,
        7.86642608, 7.81261455, 8.01682897, 6.12356659, 7.16481631, 6.87835111,
        7.57695559, 7.00287959, 7.01906241, 8.5154656, 7.58091056, 8.20254192,
        7.1648104, 7.95076475, 7.5852123, 7.14410815, 7.55442751, 6.83719233,
        6.80732498, 7.85252482, 7.43360137, 7.28726571, 7.57108561, 8.30677981
    ],
    "class": [
        'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1',
        'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1',
        'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1', 'C1',
        'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2',
        'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2',
        'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2', 'C2',
        'C2', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3',
        'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3',
        'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3', 'C3'
    ]
}

# Готовим наши данные
X = np.array(data['X'])  # Признаки
Y_classes = np.array(data['class'][:len(X)])  # Оставляем столько классов, сколько есть признаков

# Нормализуем признаки
X_min = np.min(X)
X_max = np.max(X)
X_norm = (X - X_min) / (X_max - X_min)

# Преобразуем классы в числовой вид
unique_classes = sorted(list(set(Y_classes)))
class_mapping = {cls: i for i, cls in enumerate(unique_classes)}
Y_numeric = np.array([class_mapping[c] for c in Y_classes])

# Кодируем метки классов с помощью One-Hot Encoder
onehot_encoder = OneHotEncoder(sparse_output=False)
Y_encoded = onehot_encoder.fit_transform(Y_numeric.reshape(-1, 1))

# Разделяем данные на тренировочную и тестовую выборки
X_train, X_test, Y_train, Y_test = train_test_split(
    X_norm.reshape(-1, 1), Y_encoded, test_size=0.2, random_state=42
)

# Создаем и компилируем модель
model = Sequential([
    Dense(units=8, activation='relu', input_shape=(1,)),  # Входной слой
    Dense(units=8, activation='relu'),                    # Скрытый слой
    Dense(units=Y_encoded.shape[1], activation='softmax')  # Выходной слой
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Обучение модели
history = model.fit(X_train, Y_train, epochs=100, validation_data=(X_test, Y_test))

# Оценка качества модели
score = model.evaluate(X_test, Y_test)
print(f"Точность на тестовых данных: {score[1]*100:.2f}%")

# Визуализация результатов
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Тренировочная точность')
plt.plot(history.history['val_accuracy'], label='Тестовая точность')
plt.title('Динамика точности')
plt.xlabel('Эпоха')
plt.ylabel('Точность')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Тренировочные потери')
plt.plot(history.history['val_loss'], label='Тестовые потери')
plt.title('Динамика потерь')
plt.xlabel('Эпоха')
plt.ylabel('Потери')
plt.legend()

plt.show()
```
